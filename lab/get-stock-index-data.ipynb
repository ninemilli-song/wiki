```python
import pandas as pd

def get_daily(code, start='19900101', end=''):
    url_mod = 'http://quotes.money.163.com/service/chddata.html?code=%s&start=%s&end=%s'
    url = url_mod%(code, start, end)
    df = pd.read_csv(url, encoding='gb2312')
    return df
```

```python
# 获取上证指数
df = get_daily('0000001')

print(df)
```

# 接口说明：

http://quotes.money.163.com/service/chddata.html?code=0000300&start=20151219&end=20201108&fields=TCLOSE;HIGH;LOW;TOPEN;LCLOSE;CHG;PCHG;VOTURNOVER

# 参数：

## code: 标的代码 

0000300 = 0(sh) + 000300(沪深300) 1399006 = 1(sz) + 399006(创业板指)

## start/end: 起止日期

## fields: 字段

| 字段名 | 描述 |
|-------|------|
| TCLOSE | 收盘价 |
| HIGH | 最高价 |
| LOW | 最低价 |
| CHG | 涨跌额 |
| PCHG | 涨跌幅 |
| TURNOVER | 换手率 |
| VOTURNOVER | 成交量 |
| VATURNOVER | 成交金额 |
| TCAP | 总市值 |
| MCAP | 流通市值 |
